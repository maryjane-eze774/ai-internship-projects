```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_groq import ChatGroq

import os
from dotenv import load_dotenv

load_dotenv()

print("API key loaded!" if os.getenv("GROQ_API_KEY") else "ERROR: Key not found!")

# Load PDF
loader = PyPDFLoader(r"C:\Users\USER\OneDrive - hull.ac.uk\Jane project Ai intern\Reports.pdf")
docs = loader.load()

# Split into chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.split_documents(docs)

# Embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Vector store
db = Chroma.from_documents(chunks, embeddings)

# Retriever
retriever = db.as_retriever()

# FIX: Use a supported Groq model
llm = ChatGroq(model="llama-3.3-70b-versatile")

# Combine docs manually
def combine_docs(docs):
    return "\n\n".join([d.page_content for d in docs])

# RAG function
def rag(question):
    docs = retriever.invoke(question)
    context = combine_docs(docs)
    prompt = f"Use the following context to answer the question.\n\nContext:\n{context}\n\nQuestion: {question}\nAnswer:"
    return llm.invoke(prompt)

# Ask a question
response = rag("Summarise the main findings of the report.")
print(response.content)
```
