```python
import os
from dotenv import load_dotenv

load_dotenv()

print("API key loaded!" if os.getenv("GROQ_API_KEY") else "ERROR: Key not found!")

from langchain_groq import ChatGroq
from langchain_core.tools import tool
from langgraph.prebuilt import create_react_agent

# Set up the LLM
llm = ChatGroq(model="llama-3.3-70b-versatile")

# Create tools for the agent
@tool
def get_housing_info(query: str) -> str:
    """Get information about housing market trends and data."""
    housing_data = {
        "average_price": "The average UK house price is £285,000 as of 2024",
        "trend": "UK house prices have risen 3.2% year on year",
        "first_time_buyer": "Average first time buyer deposit is £53,000",
        "hotspot": "The fastest growing regions are the East Midlands and Yorkshire",
        "forecast": "House prices are forecast to grow 4% in 2025"
    }
    for key, value in housing_data.items():
        if key in query.lower() or any(word in query.lower() for word in value.lower().split()):
            return value
    return "The UK housing market is complex with regional variations in price and demand."

@tool
def calculate_affordability(annual_salary: str) -> str:
    """Calculate housing affordability based on annual salary."""
    try:
        salary = float(annual_salary)
        max_mortgage = salary * 4.5
        min_deposit = max_mortgage * 0.1
        return f"On a salary of £{salary:,.0f}, you could borrow up to £{max_mortgage:,.0f} and would need a minimum deposit of £{min_deposit:,.0f}"
    except:
        return "Please provide a valid annual salary amount"

# Create the agent
tools = [get_housing_info, calculate_affordability]
agent = create_react_agent(llm, tools)

print("Agent ready!")
```

```python
def chat(question):
    response = agent.invoke({"messages": [{"role": "user", "content": question}]})
    print("Question:", question)
    print("Answer:", response["messages"][-1].content)
    print("-" * 50)

# Test your agent with these questions
chat("What is the average house price in the UK?")
chat("Can someone with a salary of 35000 afford a house?")
chat("What are the housing market trends?")
```
